```python
# Importar el csv

import pandas as pd
import matplotlib.pylab as plt
import numpy as np
url = "Iris_data_sample.csv"
df = pd.read_csv(url, header=None)
```

```python
df.head()
df.tail()
headers = ["0", "1", "2", "3", "4", "5"]
```

```python
# Imprimir la tabla

df.columns = headers
df.head(6)
```

```python
# Reemplazar los headers por nuevos nombres

df = df.rename(columns = {'SepalLength':'SepalLengthCm'}, inplace = False)
df.head(6)
```

```python
# Eliminar la primer fila

df.drop(0,axis=0,inplace=True)
df.head(6)
```

```python
# Reemplazar los valores ?? por NaN

df.replace("??", np.nan, inplace=True)
df.head(6)
```

```python
# Reemplazar los valores ### por NaN

df.replace("###", np.nan, inplace=True)
df.head(6)
```

```python
# Imprimir toda la tabla 

print(df.to_string())
```

```python
# Imprimir la tabla con False o true dependiendo si los espacios contienen NaN

missing_data = df.isnull()
missing_data.head(150)
```

```python
# Imprimir los campos de una forma mas legible

for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
# Sacar promedios

avg_NaN = df["NaN"].astype("float").mean(axis=0)
print("Average of NaN:", avg_NaN)
```

```python
avg_SepalLengthCm = df["SepalLengthCm"].astype("float").mean(axis=0)
print("Average of SepalLengthCm:", avg_SepalLengthCm)
```

```python
avg_SepalWidthCm = df["SepalWidthCm"].astype("float").mean(axis=0)
print("Average of SepalWidthCm:", avg_SepalWidthCm)
```

```python
avg_PetalLengthCm = df["PetalLengthCm"].astype("float").mean(axis=0)
print("Average of PetalLengthCm:", avg_PetalLengthCm)
```

```python
avg_PetalWidthCm = df["PetalWidthCm"].astype("float").mean(axis=0)
print("Average of PetalWidthCm:", avg_PetalWidthCm)
```
