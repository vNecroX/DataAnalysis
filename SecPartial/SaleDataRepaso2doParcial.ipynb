```python
count, bin_edges = np.histogram(dfT, 15)

dfT.plot(
            kind='hist',
            figsize=(10, 6),
            bins=15,
            alpha=0.6,
            xticks=bin_edges,
            color=['coral', 'darkslateblue']
         )

plt.title('Histogram of Units')
plt.ylabel("Number of units")
plt.xlabel("Number of units amount")
plt.show()
```

```python
count, bin_edges = np.histogram(dfT, 15)
xmin = bin_edges[0] - 10
xmax = bin_edges[-1] + 10

dfT.plot(
            kind='hist',
            figsize=(10, 6),
            bins=15,
            alpha=0.6,
            xticks=bin_edges,
            color=['coral', 'darkslateblue'],
            stacked=True,
            xlim=(xmin, xmax)
         )

plt.title('Histogram of Units')
plt.ylabel("Number of units")
plt.xlabel("Number of units amount")
plt.show()
```
